```python
def make_model(dir_, model_name, hden, nhi):
    full_model_name = '{0}_{1:.2f}_{2:.2f}'.format(model_name, hden, nhi)
    c_input = pc.CloudyInput('{0}{1}'.format(dir_, full_model_name))
    c_input.init_all()
    c_input.set_radius(r_in = 16)

    options = (f'hden {hden} log', 
           'metals grains -0.7 log',
           'abundances ISM', 
           'table ISM 10',
           'CMB redshift 0.0', 
           'table star tlusty Bstar 3-dim 20000 4.0 -0.7',
           'luminosity 3 solar',
           'cosmic rays background',
           f'stop neutral column density {nhi}',
           'stop temperature off',
           'set trim -20',
           'iterate to convergence',
           'save species column densities last "columns.txt" all'
          )

    c_input.set_other(options)
    c_input.print_input(to_file = True, verbose = False)
```

```python
dir_ = work_dir
pc.print_make_file(dir_ = dir_)
pc.log_.level = 2

model_name = "SMC"
```

```python
tab_hden = np.linspace(-1, 2, 4)
tab_nhi = np.linspace(19.5, 22.0, 6)
```

```python
# defining the models and writing input files
for hden in tab_hden:
    for nhi in tab_nhi:
        make_model(dir_, model_name, hden, nhi)
```

```python
# Running the models using the makefile and n_proc processors
n_proc = 24
# Take, care, this will run 20 cloudy models on 8 processors! May take some time. 
# If you run all the models togeter (n_proc = 20), you will need 10 Go RAM.
pc.run_cloudy(dir_ = dir_, n_proc = n_proc, model_name = model_name, use_make = True)
```

```python
Ms = pc.load_models('{0}{1}'.format(dir_, model_name), read_grains = False)
print(Ms)
```

```python
all_results = []

for hden in tab_hden:
    for nhi in tab_nhi:
        if hden == 0 and nhi == 21:
            continue
        if hden == 0 and nhi == 22:
            continue
        file_name = f'{model_name}_{hden:.2f}_{nhi:.2f}columns.txt'
        file_path = os.path.join(dir_, file_name)

        with open(file_path, "r") as f:
            header = f.readline().strip().split("\t")
            header[0] = "H"
            rows = [[float(x) for x in line.split("\t")] for line in f if line.strip()]

        df = pd.DataFrame(rows, columns = header)
        df["log10_2H2_over_H"] = np.log10(2) + np.log10(df["H2"]) - np.log10(df["H"])
        df["IC"] = np.log10((df["Zn"] + df["Zn+"] + df["Zn+2"] + df["Zn+3"] + df["Zn+4"] + df["Zn+5"] + df["Zn+6"] + df["Zn+7"] + df["Zn+8"] + df["Zn+9"]) / (df["H"] + df["H+"] + 2 * df["H2"] + df["H-"])) - np.log10(df["Zn+"] / (df["H"] + 2 * df["H2"]))
        df["Zn_not_+/Zn+"] = np.log10((df["Zn"] + df["Zn+2"] + df["Zn+3"] + df["Zn+4"] + df["Zn+5"] + df["Zn+6"] + df["Zn+7"] + df["Zn+8"] + df["Zn+9"]) / df["Zn+"])
        df["Zn_not_+_or_neutral/Zn+"] = np.log10((df["Zn+2"] + df["Zn+3"] + df["Zn+4"] + df["Zn+5"] + df["Zn+6"] + df["Zn+7"] + df["Zn+8"] + df["Zn+9"]) / df["Zn+"])
        df["hden"] = hden
        df["nhi"] = np.log10(df["H"])

        result = df[["log10_2H2_over_H", "nhi", "IC", "hden", "Zn_not_+/Zn+", "Zn_not_+_or_neutral/Zn+"]]
        all_results.append(result)

full_table = pd.concat(all_results, ignore_index = True)
# print(full_table)
```

```python
sc = plt.scatter(full_table["nhi"], full_table["log10_2H2_over_H"], c = full_table["IC"], cmap = "afmhot")

cbar = plt.colorbar(sc)
cbar.set_label("IC(ZnII)")

plt.xlabel(r"$\log_{10}(N_{\rm H})$")
plt.ylabel(r"$\log_{10}\left(f_{H_2}\right)$")

plt.show()
```

```python
sc = plt.scatter(full_table["nhi"], full_table["log10_2H2_over_H"], c = full_table["Zn_not_+/Zn+"], cmap = "afmhot")

cbar = plt.colorbar(sc)
cbar.set_label("Log(Zn not in ZnII/ZnII)")

plt.xlabel(r"$\log_{10}(N_{\rm H})$")
plt.ylabel(r"$\log_{10}\left(f_{H_2}\right)$")

plt.show()
```

```python
sc = plt.scatter(full_table["nhi"], full_table["log10_2H2_over_H"], c = full_table["Zn_not_+_or_neutral/Zn+"], cmap = "afmhot")

cbar = plt.colorbar(sc)
cbar.set_label("Log(Zn not in ZnI or ZnII/ZnII)")

plt.xlabel(r"$\log_{10}(N_{\rm H})$")
plt.ylabel(r"$\log_{10}\left(f_{H_2}\right)$")

plt.show()
```
